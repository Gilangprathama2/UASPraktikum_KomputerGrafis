```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Stocks.csv')
df
```

**A. Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histogram**

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price ($)'], kde=True, bins=13, edgecolor='w', color="green")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price ($)'], kde=True, bins=11, edgecolor='w', color="red")
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss ($)'], kde=True, bins=13, edgecolor='w', color="blue")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Gain/Loss')

plt.tight_layout()
plt.show()
```

**b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
data = df.sort_values(by="Dividend/Shares", ascending=False)

top_5 = data.head(5)
#plot data
plt.figure(figsize=(10, 5))
sns.barplot(x="Stock Name", y="Dividend/Shares", data=top_5, color="green")
plt.title("Top 5 Stocks Share Tertinggi")
plt.show()
```

**c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi**

```python
data = df.sort_values(by="Market Value ($)", ascending=False)

top_5 = data.head(5)
#plot data
plt.figure(figsize=(10, 5))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5, color="orange")
plt.title("Top 5 Market Value ($) Tertinggi")
plt.show()
```

**d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi**

```python
data = df.sort_values(by="Annual Yield (%)", ascending=False)

top_7 = data.head(7)
#plot data
plt.figure(figsize=(10, 5))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7, color="red")
plt.title("Top 7 Annual Yield ($) Tertinggi")
plt.show()
```
